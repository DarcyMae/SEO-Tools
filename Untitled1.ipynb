```python
!pip install langdetect
!pip install gradio
!pip install deep_translator
import os
from dotenv import load_dotenv
import tensorflow as tf
from googletrans import Translator
from tensorflow.keras.preprocessing.sequence import pad_sequences
import pickle
import anthropic
import gradio as gr
import re
import numpy as np
from langdetect import detect
from deep_translator import GoogleTranslator
from collections import Counter
import matplotlib.pyplot as plt
import io
import csv
from datetime import datetime

```

```python
# Try to import anthropic, and handle potential import error
try:
    import anthropic
except ImportError:
    print("Error: anthropic library not found. Please install it using 'pip install anthropic'")
    raise

# Load environment variables from .env file
load_dotenv()

# Get the API key from environment variables
ClaudeAPI_Key = os.getenv('ClaudeAPI_Key')

if not ClaudeAPI_Key:
    raise ValueError("ClaudeAPI_Key not found in .env file. Please make sure it's set correctly.")

# Load your pre-trained model and tokenizer here
try:
    model = tf.keras.models.load_model('SecondRun.keras')
    with open('tokenizer.pickle', 'rb') as handle:
        tokenizer = pickle.load(handle)
except Exception as e:
    print(f"Error loading model or tokenizer: {e}")
    print("Proceeding without the model. Some functionalities may be limited.")
    model = None
    tokenizer = None

# Initialize the Anthropic client
try:
    client = anthropic.Anthropic(api_key=ClaudeAPI_Key)
except Exception as e:
    print(f"Error initializing Anthropic client: {e}")
    client = None

# Initialize the Google Translator
translator = GoogleTranslator(source='auto', target='en')

def translate_if_needed(text, target_language="en"):
    try:
        source_language = detect(text)
        if source_language != target_language:
            return translator.translate(text)
    except Exception as e:
        print(f"Error in translation: {e}")
    return text

# Rest of your code remains the same
```

```python

# Cache for Claude API calls
sentiment_cache = {}
response_cache = {}

def analyze_sentiment(review):
    try:
        sequence = tokenizer.texts_to_sequences([review])
        padded_sequence = pad_sequences(sequence, maxlen=50)
        lstm_prediction = model.predict(padded_sequence)[0][0]
        lstm_sentiment = "positive" if lstm_prediction > 0.5 else "negative"
        confidence = lstm_prediction if lstm_sentiment == "positive" else 1 - lstm_prediction
        return lstm_sentiment, confidence
    except Exception as e:
        print(f"Error in LSTM sentiment analysis: {e}")
        return "unknown", 0.0

def get_claude_sentiment(review):
    if review in sentiment_cache:
        return sentiment_cache[review]

    try:
        initial_message = f'''Analyze the sentiment of this Google review:
        <google_review>
        {review}
        </google_review>
        Determine if it is positive or negative. Consider the overall tone, specific praise or complaints, and any star rating. 
        Respond with only two lines:
        1. The sentiment (positive or negative)
        2. A confidence score between 0 and 1'''

        messages = [
            {"role": "user", "content": [{"type": "text", "text": initial_message}]}
        ]

        response = client.messages.create(
            model="claude-3-5-sonnet-20240620",
            max_tokens=100,
            temperature=0,
            messages=messages
        )
        
        claude_analysis = response.content[0].text if isinstance(response.content, list) else response.content
        
        lines = claude_analysis.strip().split('\n')
        if len(lines) >= 2:
            claude_sentiment = lines[0].lower()
            confidence = float(lines[1])
        else:
            claude_sentiment = "unknown"
            confidence = 0.0

        result = (claude_sentiment, confidence, claude_analysis)
        sentiment_cache[review] = result
        return result
    except Exception as e:
        print(f"Error in Claude sentiment analysis: {e}")
        return "unknown", 0.0, ""

def generate_response(review, sentiment, tone="neutral", additional_context=""):
    cache_key = (review, sentiment, tone, additional_context)
    if cache_key in response_cache:
        return response_cache[cache_key]

    try:
        prompt = f'''Respond to this Google review:
        <google_review>
        {review}
        </google_review>

        Sentiment: {sentiment}
        Tone: {tone}

        Additional context:
        {additional_context}

        Craft a response that addresses the review appropriately based on its sentiment and the provided tone. 
        For negative reviews, address concerns and invite the customer to give the business another chance.
        Write your response inside <response> tags.'''

        messages = [
            {"role": "user", "content": [{"type": "text", "text": prompt}]}
        ]

        response = client.messages.create(
            model="claude-3-5-sonnet-20240620",
            max_tokens=1000,
            temperature=0,
            messages=messages
        )

        claude_response = response.content[0].text if isinstance(response.content, list) else response.content

        response_match = re.search(r'<response>(.*?)</response>', claude_response, re.DOTALL)
        final_response = response_match.group(1).strip() if response_match else "No response found."

        response_cache[cache_key] = final_response
        return final_response
    except Exception as e:
        print(f"Error generating response: {e}")
        return "Error generating response. Please try again."

def detect_language(text):
    try:
        return detect(text)
    except:
        return "en"  # default to English if detection fails

def translate_if_needed(text, target_language="en"):
    source_language = detect_language(text)
    if source_language != target_language:
        return translator.translate(text, src=source_language, dest=target_language).text
    return text

def extract_keywords(text):
    words = re.findall(r'\w+', text.lower())
    return [word for word in words if len(word) > 3]

def visualize_sentiment(sentiments):
    labels = ['Positive', 'Negative', 'Neutral']
    sizes = [sentiments.count('positive'), sentiments.count('negative'), sentiments.count('neutral')]
    colors = ['#66b3ff', '#ff9999', '#99ff99']
    plt.figure(figsize=(8, 6))
    plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
    plt.axis('equal')
    plt.title('Sentiment Distribution')
    buf = io.BytesIO()
    plt.savefig(buf, format='png')
    buf.seek(0)
    return buf

def process_reviews(reviews):
    results = []
    sentiments = []
    keywords = []

    for review in reviews:
        lstm_sentiment, lstm_confidence = analyze_sentiment(review)
        claude_sentiment, claude_confidence, _ = get_claude_sentiment(review)
        
        agreement = "Agree" if lstm_sentiment == claude_sentiment else "Disagree"
        final_sentiment = claude_sentiment if claude_confidence > lstm_confidence else lstm_sentiment
        sentiments.append(final_sentiment)
        
        translated_review = translate_if_needed(review)
        keywords.extend(extract_keywords(translated_review))
        
        result = {
            'review': review,
            'lstm_sentiment': lstm_sentiment,
            'lstm_confidence': lstm_confidence,
            'claude_sentiment': claude_sentiment,
            'claude_confidence': claude_confidence,
            'final_sentiment': final_sentiment,
            'agreement': agreement
        }
        results.append(result)

    sentiment_distribution = visualize_sentiment(sentiments)
    keyword_freq = Counter(keywords).most_common(10)

    return results, sentiment_distribution, keyword_freq

def export_results(results):
    filename = f"review_analysis_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
    with open(filename, 'w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=results[0].keys())
        writer.writeheader()
        writer.writerows(results)
    return filename

def process_single_review(review, tone):
    if not review.strip():
        return "Please enter a review.", None, None, None

    translated_review = translate_if_needed(review)
    lstm_sentiment, lstm_confidence = analyze_sentiment(translated_review)
    claude_sentiment, claude_confidence, claude_analysis = get_claude_sentiment(translated_review)
    
    agreement = "Agree" if lstm_sentiment == claude_sentiment else "Disagree"
    final_sentiment = claude_sentiment if claude_confidence > lstm_confidence else lstm_sentiment
    
    sentiment_output = f"LSTM: {lstm_sentiment} (confidence: {lstm_confidence:.2f})\n" \
                       f"Claude: {claude_sentiment} (confidence: {claude_confidence:.2f})\n" \
                       f"Final sentiment: {final_sentiment}\n" \
                       f"Agreement: {agreement}"
    
    response = generate_response(translated_review, final_sentiment, tone)
    
    return sentiment_output, response, gr.update(visible=True), gr.update(visible=True)

def batch_process_reviews(reviews, tone):
    reviews_list = reviews.split('\n')
    results, sentiment_distribution, keyword_freq = process_reviews(reviews_list)
    
    output = "Batch Processing Results:\n\n"
    for result in results:
        output += f"Review: {result['review']}\n"
        output += f"Final Sentiment: {result['final_sentiment']}\n"
        output += f"Agreement: {result['agreement']}\n\n"
    
    export_file = export_results(results)
    
    return output, sentiment_distribution, gr.update(visible=True, value=f"Results exported to {export_file}")

with gr.Blocks() as demo:
    gr.Markdown("# Advanced Google Review Analyzer and Responder")
    
    with gr.Tab("Single Review"):
        review_input = gr.Textbox(label="Google Review")
        tone_input = gr.Radio(["formal", "casual", "empathetic"], label="Response Tone", value="neutral")
        analyze_button = gr.Button("Analyze Review")
        
        sentiment_output = gr.Textbox(label="Sentiment Analysis")
        response_output = gr.Textbox(label="Generated Response", visible=False)
        feedback_input = gr.Radio(["Thumbs Up", "Thumbs Down"], label="Feedback", visible=False)
    
    with gr.Tab("Batch Processing"):
        batch_input = gr.Textbox(label="Enter multiple reviews (one per line)")
        batch_tone_input = gr.Radio(["formal", "casual", "empathetic"], label="Response Tone", value="neutral")
        batch_button = gr.Button("Process Batch")
        
        batch_output = gr.Textbox(label="Batch Results")
        sentiment_plot = gr.Plot(label="Sentiment Distribution")
        export_output = gr.Textbox(label="Export Status", visible=False)
    
    analyze_button.click(
        process_single_review,
        inputs=[review_input, tone_input],
        outputs=[sentiment_output, response_output, response_output, feedback_input]
    )
    
    batch_button.click(
        batch_process_reviews,
        inputs=[batch_input, batch_tone_input],
        outputs=[batch_output, sentiment_plot, export_output]
    )

demo.launch()
```
